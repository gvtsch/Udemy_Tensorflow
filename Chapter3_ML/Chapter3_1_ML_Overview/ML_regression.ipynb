```python
import matplotlib.pyplot as plt
from tf_utils.dummyData import regression_data
```

```python
def model(x):
    m = 2.0
    b = 5.0
    return m * x + b
```

```python
if __name__=='__main__':
    x, y = regression_data()

    y_pred = model(x)

    plt.scatter(x, y)
    plt.plot(x, y_pred, color = 'red')
    plt.show()
```
