```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_boston
```

```python
if __name__=='__main__':
    dataset = load_boston()
    x = dataset.data
    y = dataset.target

    df = pd.DataFrame(x, columns=dataset.feature_names)
    df['y'] = y

    print(df.head(n=10))
    print(df.info())
    print(df.describe())

    df.hist(bins=30, figsize=(15,15))
    plt.show()
```

In dem Histogramm sieht man die 13 verschiedenen Features des Boston-Datasets.
Man sollte darauf achten, dass die Daten eines Features gleich bzw. normalverteilt sind. In einzelnen Features hier sieht man starke Ausreißer bzw. unausgewogene Verteilungen. Mit solchen Ausreißern oder "Unverteilungen" kann das Training behindert werden. Das mindert die Performance. Daten müssen dann entsprechend aufbereitet, die Güte gesteigert werden.
